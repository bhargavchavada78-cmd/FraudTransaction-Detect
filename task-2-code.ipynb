**import Required libraries,import machine learning model,import numerical operations**

```python
import numpy as np
import pandas as pd
from sklearn.metrics import accuracy_score
from sklearn.linear_model  import LogisticRegression
from sklearn.linear_model  import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
```

**Load the datasets**

```python
test_df= pd.read_csv('fraudTest.csv')
train_df = pd.read_csv('fraudTrain.csv')
```

**Drop non-important Colums**

```python
drop_colm = ['merchant','first','last','street','zip','unix_time','job','cc_num','trans_num','trans_date_trans_time','city','dob','merch_lat','merch_long']
```

```python
test_df = test_df.drop(columns=drop_colm, axis=1)
train_df = train_df.drop(columns=drop_colm, axis=1)
```

**Remove Null values**

```python
train_df = train_df.dropna(subset=['is_fraud'])
test_df = test_df.dropna(subset=['is_fraud'])
```

```python
train_df['is_fraud'].isna().sum()
test_df['is_fraud'].isna().sum()
```

**Separate features (X) and Target variable (Y)**

**'is_fraud' is the target column**

```python
X_train = train_df.drop(columns='is_fraud', axis=1)
Y_train = train_df['is_fraud']
```

```python
X_test = test_df.drop(columns='is_fraud', axis=1)
Y_test = test_df['is_fraud']
```

**Convert categorical variable into numerical form using one-hot encoding**

```python
X_train = pd.get_dummies(X_train)
X_test = pd.get_dummies(X_test)
```

**Align training   a  nd testing datasets to ensure same feature columns**

```python
X_train,X_test =X_train.align(X_test,join="left",axis=1,fill_value=0)
```

**Scale the feature values to improve model performance**

```python
Scaler = StandardScaler()
X_train_Scaled = Scaler.fit_transform(X_train)
X_test_Scaled = Scaler.transform(X_test)
```

**chack model before train**

```python
print(X_train_Scaled.shape)
print(Y_train.shape)
print(Y_train.isna().sum())
```

**Initialize the LogisticRegression**

```python
lr = LogisticRegression(max_iter=1000,class_weight='balanced')
lr.fit(X_train_Scaled,Y_train)
```

**Predict fraud on test data**

```python
y_pred = lr.predict(X_test_Scaled)
```

**Evaluate model performance**

```python
print(classification_report(Y_test,y_pred))
```

**Initialize the RandomForestClassifier**

```python
rf = RandomForestClassifier(n_estimators=50,max_depth=50,class_weight='balanced')
rf.fit(X_train_Scaled,Y_train)
```

**Predict fraud on test data**

```python
y_pred = rf.predict(X_test_Scaled)
```

**Evaluate model performance**

```python
print(classification_report(Y_test,y_pred))
```

**Intialize the Decision Tree Classifier**

```python
dt = DecisionTreeClassifier(max_depth = 55,class_weight='balanced',min_samples_split=100,min_samples_leaf=50,random_state=42)
dt.fit(X_train_Scaled,Y_train)
```

**Predict fraud on test data**

```python
y_pred = dt.predict(X_test_Scaled)
```

**Evaluate model performance**

```python
print(classification_report(Y_test,y_pred))
```

**Take input from user**

```python
user_input = {
    'amt': 120.50,
    'lat': 40.7128,
    'long': -74.0060,
    'city_pop': 800000,
    'category': 'food_dining',
    'gender': 'F',
    'state': 'NY'
}
```

**Convert input in DataFrame**

```python
user_df = pd.DataFrame([user_input])
```

```python
user_df = pd.get_dummies(user_df)

user_df = user_df.reindex(columns=X_train.columns, fill_value=0)
```

```python
user_scaled = Scaler.transform(user_df)
```

**Predict Final Result**

```python
prediction = dt.predict(user_scaled)

if prediction[0] == 1:
    print("⚠️ Fraud Transaction Detected")
else:
    print("✅ Legitimate Transaction")
```
